```
#Achico un poco para calcular
z=datos1[datos1['Country']=='France']
z.head()
```

```
z.shape
```

```
z.info()
```

```
z['Description']=z['Description'].str.strip()#Tengo que eliminar espacios
z['Description']
```

```
h=z.groupby(['InvoiceNo', 'Description'])
h
                                             
```

```
fin=h['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
fin.head()
```

```
#Encode 0 y 1
def hotencode(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

fin_fin = fin.applymap(hotencode)
fin_fin.head()
```

```
fin_fin['POSTAGE'].head()# Vi que sugieren sacarlo 
```

```
fin_fin.drop('POSTAGE', inplace=True, axis=1)
# from sklearn.preprocessing import OneHotEncoder
#fin_fin=fit_transform(fin,handle_unknown='ignore')
```

```
#Empezamos con las reglas de asociación
frecuencia_item = apriori(fin_fin, min_support=0.07, use_colnames=True)
print(frecuencia_item.head())

```

```
regla = association_rules(frecuencia_item, metric="lift", min_threshold=1)
regla.head()
```
